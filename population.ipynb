```python
import sqlite3

# Create a SQL connection to our SQLite database
con = sqlite3.connect("persons_40K.db")

cur = con.cursor()

# The result of a "cursor.execute" can be iterated over by row
for row in cur.execute('SELECT distinct provincia FROM person;'):
    print(row)

# Be sure to close the connection
con.close()
```

```python
import sqlite3
import pandas as pd

# Create a SQL connection to our SQLite database
con = sqlite3.connect("persons_40K.db")

df = pd.read_sql("select * from person", con=con)
print(df)
con.close()
```

```python
import sqlite3
import pandas as pd

# Create a SQL connection to our SQLite database
con = sqlite3.connect("persons_40K.db")

df = pd.read_sql("select provincia, count(*) as habitantes from person group by provincia;", con=con)
print(df)
con.close()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Create a SQL connection to our SQLite database
con = sqlite3.connect("persons_40K.db")

df = pd.read_sql("select provincia, count(*) as habitantes from person group by provincia;", con=con)
plt.style.use('ggplot')
ax = df.sort_values(by='habitantes',ascending=True).plot.bar(x='provincia', title ="Población",figsize=(15,10),legend=True, fontsize=12)
ax.set_xlabel("Provincia",fontsize=12)
ax.set_ylabel("Poblacion",fontsize=12)
plt.show()

con.close()
```

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Create a SQL connection to our SQLite database
con = sqlite3.connect("persons_40K.db")

df = pd.read_sql("select strftime('%Y',f_nacimiento) as any_nacimiento, count(*) as habitantes from person group by provincia;", con=con)
plt.style.use('ggplot')
ax = df.sort_values(by='any_nacimiento',ascending=True).plot.bar(x='any_nacimiento', title ="Población",figsize=(15,10),legend=True, fontsize=12)
ax.set_xlabel("Año",fontsize=12)
ax.set_ylabel("Habitantes",fontsize=12)
plt.show()

con.close()
```
